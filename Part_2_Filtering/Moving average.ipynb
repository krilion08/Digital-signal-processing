```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
def moving_average(signal, window: int=2):
    #computes moving average of a signal
    N = window - 1 #since indexing starts with 0
    filtered_signal = signal.copy()
    for i, v in enumerate(filtered_signal[N:]):
        filtered_signal[N+i] = filtered_signal[i:N+i+1].sum() / window
    return(filtered_signal)
```

```python
X = np.linspace(0, 20, 500)
noise = np.random.default_rng().normal(0., 0.1, 500)
Y = np.sin(X)
Y_noisy = Y + noise
plt.plot(Y, label='original signal')
plt.plot(Y_noisy, label='noisy')
filtered_signal = moving_average(Y_noisy, 4)
plt.plot(filtered_signal, label='filtered signal')
plt.legend()
plt.show()
```
